# Skirt

```python
df_skirt = pd.read_csv('./review_skirt.csv', index_col = 'Unnamed: 0')
cloth_skirt = pd.read_json('./final_skirt.json')
df_skirt.insert(1, 'largecategory', 5, True)
df_skirt
```

```python
cloth_skirt.columns
```

```python
cloth_skirt = cloth_skirt[['newClothId', 'clothId', 'smallCategoryMiniskirt', 'smallCategoryMidi', 'smallCategoryLongskirt',
       'colorWhite', 'colorGrey', 'colorBlack', 'colorRed', 'colorPink', 'colorOrange', 'colorIvory', 'colorYellow',
       'colorGreen', 'colorBlue', 'colorPurple', 'colorBrown', 'colorBeige', 'colorJean', 'colorPattern', 'colorOthers']]       
```

```python
df_skirt = pd.merge(df_skirt, cloth_skirt, how="left", left_on='newGoodsNo', right_on="newClothId")
df_skirt
```

```python
df_skirt.columns
```

## User 데이터 생성 (unique한 유저 정보로 수정)

```python
user_skirt = df_skirt[['userName', 'largecategory', 'userSexMen', 'userSexWomen']]
user_skirt = user_skirt.drop_duplicates(subset=['userName'])
user_skirt.sort_values('userName', inplace=True)
user_skirt.index = range(len(user_skirt))
```

```python
mo = df_skirt.groupby('userName').sum()[['month1', 'month2', 'month3', 'month4', 'month5', 'month6', 'month7', 'month8', 'month9', 'month10', 'month11', 'month12']]
mo.sort_values('userName', inplace=True)
mo.index = range(len(mo))
```

```python
cnt = df_skirt.groupby('userName').count()
cnt.sort_values('userName', inplace=True)
cnt.index = range(len(mo))
cnt = cnt[['largecategory']]
cnt.columns = ['clothCnt']
cnt
```

```python
hw = df_skirt.groupby('userName').mean()[['userHeight', 'userWeight', 'size', 'bright', 'color', 'thickness', 'weightness', 'touch', 'smallCategoryMiniskirt',
       'smallCategoryMidi', 'smallCategoryLongskirt', 'colorWhite',
       'colorGrey', 'colorBlack', 'colorRed', 'colorPink', 'colorOrange',
       'colorIvory', 'colorYellow', 'colorGreen', 'colorBlue', 'colorPurple',
       'colorBrown', 'colorBeige', 'colorJean', 'colorPattern', 'colorOthers']]
hw.sort_values('userName', inplace=True)
hw.index = range(len(hw))
```

```python
user_skirt = pd.concat([user_skirt, hw, mo, cnt], axis=1)
user_skirt
```

```python
user_skirt.columns
```

```python
user_skirt.columns = ['userName', 'largecategory', 'userMale', 'userFemale', 'userHeight',
       'userWeight', 'size', 'bright', 'color', 'thickness', 'weightness',
       'touch', 'smallCategoryMiniskirt', 'smallCategoryMidi',
       'smallCategoryLongskirt', 'colorWhite', 'colorGrey', 'colorBlack',
       'colorRed', 'colorPink', 'colorOrange', 'colorIvory', 'colorYellow',
       'colorGreen', 'colorBlue', 'colorPurple', 'colorBrown', 'colorBeige',
       'colorJean', 'colorPattern', 'colorOthers', 'month1', 'month2',
       'month3', 'month4', 'month5', 'month6', 'month7', 'month8', 'month9',
       'month10', 'month11', 'month12', 'clothCnt']
```

```python
user_skirt.insert(0, 'userId', 0, True)
```

```python
user_skirt['userId'] = range(len(user_skirt))
```

```python
user_skirt
```

```python
user_skirt.userMale = user_skirt.userMale.astype('int')
user_skirt.userFemale = user_skirt.userFemale.astype('int')
user_skirt.userHeight = user_skirt.userHeight.astype('int')
user_skirt.userWeight = user_skirt.userWeight.astype('int')
```

```python
user_skirt.to_json('user_skirt.json', orient='records', force_ascii=False)
```

## Transaction 데이터 생성(전체 리뷰인 경우로 수정)

```python
transaction_skirt = df_skirt[['userName', 'newGoodsNo', 'largecategory']]
```

```python
transaction_skirt = pd.merge(user_skirt[['userName', 'userId']], transaction_skirt, how="right", on="userName")
transaction_skirt
```

```python
transaction_skirt = transaction_skirt.drop('userName', axis=1)
transaction_skirt.columns = ['userId', 'newClothId', 'largecategory']
transaction_skirt
```

```python
b = transaction_skirt.drop_duplicates(subset=['userId', 'newClothId']).reset_index(drop=True)
b = b.sort_values(['userId', 'newClothId'])
b.index = range(len(b))
b
```

```python
a = transaction_skirt.groupby(['userId', 'newClothId']).count()
a = a.sort_values(['userId', 'newClothId'])
a.columns = ['shopCnt']
a.index = range(len(a))
a
```

```python
transaction_skirt = pd.concat([b, a], axis=1)
transaction_skirt
```

```python
transaction_skirt.to_json('transaction_skirt.json', orient='records', force_ascii=False)
```
