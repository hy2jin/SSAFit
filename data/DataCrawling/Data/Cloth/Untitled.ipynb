```python
import pandas as pd
```

```python
cloth = pd.read_json('real_final_skirt.json')
review = pd.read_json('review_skirt.json')
```

```python
cloth_id = list(cloth.newClothId.unique())
```

```python
cloth.shape
```

```python
cloth = cloth.rename(columns={'thickness': 'thick'})
```

```python
cloth.columns
```

```python
review
```

```python
re = review.loc[review.newGoodsNo.isin(cloth_id),:]
```

```python
re.shape
```

```python
re_style = re.groupby(['newGoodsNo']).mean()[['size', 'bright','color']]
```

```python
re_style.insert(0, 'newGoodsNo', re_style.index, True)
```

```python
re_style.index = range(len(re_style))
```

```python
re_style.insert(4, 'thickness', 0, True)
```

```python
re_style
```

```python
re_style.shape
```

```python
merge_data = pd.merge(cloth, re_style[['newGoodsNo', 'size','bright','color','thickness']], how='left', left_on='newClothId', right_on="newGoodsNo")
```

```python
merge_data
```

```python
merge_data = merge_data.drop('newGoodsNo', axis=1)
```

```python
merge_data['size'] = merge_data['size'].fillna(0)
merge_data['bright'] = merge_data['bright'].fillna(0)
merge_data['color'] = merge_data['color'].fillna(0)
merge_data['thickness'] = merge_data['thickness'].fillna(0)
```

```python
sum(merge_data['thickness'].isna())
```

```python
merge_data.shape
```

```python
merge_data.to_json('real_final_skirt.json', orient='records', force_ascii=False)
```
